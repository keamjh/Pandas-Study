### Step 14. 총 매출은 얼마인가요?

### Step 16. 주문당 평균 매출은 얼마인가요?

21.394231188658654

```python
# Solution 1 : groupby() 미사용
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
order_grouped = chipo.groupby(by=['order_id']).sum()
order_grouped.mean()['revenue']
```

```python
# Solution 2 : groupby() 사용
# chipo.groupby(by=['order_id']).sum().mean()
a = chipo.groupby(by=['order_id']).sum().mean()['revenue']
print("주문당 평균 매출: $" + str(np.round(a, 3)))
```
